__3. Track the original repository__

```python
%%bash
# Move to forked repo folder
cd /Users/maverickmiaow/Desktop/clone/master-git-github
# Configure the upstream
git remote add --track master upstream https://github.com/YanCheng-go/master-git-github.git
```

```python
%%bash
cd /Users/maverickmiaow/Desktop/clone/master-git-github

# Sync the repository
git fetch upstream
```

```python
%%bash
cd /Users/maverickmiaow/Desktop/clone/master-git-github

git checkout master
```

```python
%%bash
cd /Users/maverickmiaow/Desktop/clone/master-git-github
# Rewrite your master branch so that any commits of yours that
# aren't already in upstream/master are replayed on top of that
# other branch:
git rebase upstream/master 
```

```python
%%bash
cd /Users/maverickmiaow/Desktop/clone/master-git-github

git status
```

```python
%%bash
cd /Users/maverickmiaow/Desktop/clone/master-git-github


git rebase upstream/master
```

```python
%%bash
cd /Users/maverickmiaow/Desktop/clone/master-git-github

git push origin master
```

```python
%%bash
cd /Users/maverickmiaow/Desktop/clone/master-git-github

git log  --pretty=format:"%h - %an, %ar : %s"  --graph
```
